```python
prop = df.drop(['area_type','society','availability'], axis=1)
```

```python
prop.head()
```

```python
prop.shape
```

```python
prop['location'].value_counts()
```

```python
prop['location']=prop['location'].fillna('Electronic City')
```

```python
prop['size'].value_counts()
```

```python
prop['size']=prop['size'].fillna('2 BHK')
```

```python
prop['bath'].value_counts()
```

```python
prop['bath'] = prop['bath'].fillna(prop['bath'].median())
```

```python
prop['balcony_was_missing'] = prop['balcony'].isna()
prop['balcony'] = prop['balcony'].fillna(2)
```

```python
prop.isna().sum()
```

```python
prop['size'] = prop['size'].replace(['1 RK'],'0 BRK')
```

```python
prop['No_of_bedrooms'] = prop['size'].str.split().str.get(0).astype(int)
prop.head()
```

```python
prop[prop.No_of_bedrooms>20]
```

```python
prop = prop.drop([1718,4684])
```

```python
prop['total_sqft'].value_counts()
```

```python
def convert_range(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
prop['total_sqft'] = prop['total_sqft'].apply(convert_range)
```

```python
prop['price_per_sqft'] = prop['price']*100000/prop['total_sqft']
```

```python
prop.drop('size',inplace=True,axis=1)
```

```python
prop['location'] = prop['location'].apply(lambda x: x.strip())
loc_num = prop.groupby('location')['location'].agg('count').sort_values(ascending=False)
loc_num
```

```python
low = loc_num[loc_num<=10]
low
```

```python
prop.location = prop.location.apply(lambda x: 'other' if x in low else x)
```

```python
temp = pd.get_dummies(prop.location)
df = pd.concat([prop.drop('location',axis=1),temp],axis=True)
df = df.fillna(0)
df.head()
```

```python
#data preprocessing
X = df.drop(['price'], axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```
